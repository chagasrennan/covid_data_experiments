```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://data.cdc.gov/resource/y5bj-9g5w.csv',
                 parse_dates=['week_ending_date','time_period'])
```

```python
df.age_group.unique()
```

```python
aux = df.groupby(['week_ending_date','age_group','number_of_deaths']).sum().reset_index()

aux.pivot_table(values=['number_of_deaths'],
                index=['week_ending_date'],
                columns=['age_group']).plot(subplots=True,layout=(3,1), figsize=(12,8));
```
